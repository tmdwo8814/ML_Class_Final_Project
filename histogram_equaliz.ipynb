```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt

def histogram_equalization(image):
    image_pil = transforms.ToPILImage()(image)

    image_equalized = transforms.functional.equalize(image_pil)

    image_tensor_equalized = transforms.ToTensor()(image_equalized)

    return image_tensor_equalized

image_path = "data/test/adenocarcinoma/000108 (3).png"
image = Image.open(image_path).convert("L")  

image_tensor = transforms.ToTensor()(image)

image_equalized = histogram_equalization(image_tensor)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_tensor[0], cmap="gray")
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(image_equalized[0], cmap="gray")
plt.title("Equalized Image")

plt.show()
```
